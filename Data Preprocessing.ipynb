```python
import pandas as pd
```

```python
#read crew.csv
df_crew = pd.read_csv('crew.csv')
print(df_crew.shape)
df_crew.head()
```

```python
#Read movie.csv
df_movie = pd.read_csv('movie.csv')
print(df_movie.shape)
df_movie.head()
```

```python
#Filter movie
#df_movie with >10000 [imdb_votecount] & production_country contains US 

df_movie_f1 = df_movie.loc[(df_movie['imdb_votecount'] > 10000) & (df_movie['product_country'].str.contains('US'))]

#reset index
df_movie_f1.reset_index(drop=True,inplace=True)
print(df_movie_f1.shape)
df_movie_f1
```

```python
#create mov_list from df_movie_f1 column ['tmdb_id']
mov_list = df_movie_f1['tmdb_id'].tolist()
```

```python
#Filter crew data that is in mov_list only
df_crew_f1 = df_crew.loc[df_crew['movie_id'].isin(mov_list)]

#reset index
df_crew_f1.reset_index(drop=True,inplace=True)
```

```python
#display unique job in df
df_crew_f1.job.unique()
```

```python
#filter crew with specific jobs
job_list = ['Actor', 'Director', 'Producer', 'Executive Producer']
df_crew_f2 = df_crew_f1.loc[(df_crew_f1['job'].isin(job_list))]

#reset index
df_crew_f2.reset_index(drop=True,inplace=True)
```

```python
#Check unique name
print(len(df_crew_f2.name.unique()))
df_crew_f2.head()
```

```python
#create ['actor'] column that encode ['name'] into index starting from 0
df_crew_f2['actor'] = df_crew_f2.name.astype('category').cat.codes
```

```python
#Create ['movie'] column that encode ['movie_id'] into index starting from 0
df_crew_f2['movie'] = df_crew_f2.movie_id.astype('category').cat.codes
```

```python
df_crew_f2
```

```python
#check len(actor) == len(name)
print('check len actor == name:',len(df_crew_f2['actor'].unique()) == len(df_crew_f2['name'].unique()))
print('len actor =:', len(df_crew_f2['actor'].unique()))
print()
#check len(movie) == len(movie_id)
print('check len movie == movie_id:',len(df_crew_f2['movie'].unique()) == len(df_crew_f2['movie_id'].unique()))
print('len movie =:', len(df_crew_f2['movie'].unique()))
```

```python
df_crew_merged = pd.merge(df_crew_f2, df_movie_f1[['tmdb_id','title']], left_on='movie_id', right_on='tmdb_id', how = 'left')
```
